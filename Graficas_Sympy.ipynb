Importa la libreria <font color='red'>sympy.plotting</font>, es un módulo dentro de SymPy que se utiliza para crear gráficos de funciones matemáticas.

```python
from sympy.plotting import plot
```

Grafica $y(x)=x^2$ utlizando solamente el plot

```python
plot(x**2)
```

Grafica $y(x)=x^2$ pero de [-1,1] en el eje x, igual solo con el plot

```python
plot(x**2,(x,-1,1))
```

La primera funciòn grafica $y(x)=e^x$, y la segunda lo mismo pero desde [0,2], con plot

```python
plot(exp(x))
plot(exp(x),(x,0,2))
```

Grafica $y(x)=x,y(x)=x^2,y(x)=x^3$

```python
plot(x,x**2,x**3, (x,-2,2))
```

### Graficas 3D

```python
from sympy.plotting import plot3d
```

```python
x,y=symbols('x y')
```

```python
plot3d(x*y)
```

```python
plot3d(x*y, (x,-5,5),(y,-5,5))
```

```python
plot3d(x*y, -x*y, (x,-5,5),(y,-5,5))
```

```python
plot3d((x**2 + y**2, (x,-5,5),(y,-5,5)),(x*y,(x,-3,3),(y,-3,3)))
```

```python
from sympy.plotting import plot3d_parametric_line
```

```python
plot3d_parametric_line(cos(s),sin(s),s)
```

```python
plot3d_parametric_line((cos(s),sin(s),s,(s,-5,5)),(sin(s),s**2,s,(s,-5,5)))
```

```python
from sympy.plotting import plot3d_parametric_surface
```

```python
u, v =symbols('u v')
```

```python
plot3d_parametric_surface(cos(u+v),sin(u-v),u-v)
```

```python
plot3d_parametric_surface(cos(v)*cos(u),cos(v)*sin(u),sin(v))
```

Funciones implícitas

```python
from sympy import plot_implicit, symbols, Eq, And
```

```python
x, y = symbols(' x y')
```

```python
p1 = plot_implicit(Eq(x**2 + y**2, 5))
```

```python
p6 = plot_implicit(y > x**2)
```

```python
p6 = plot_implicit(And(y > x, y > -x))
```
