$X^{''}-\lambda X=0 \qquad X(0)=X(1)=0$


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, lam = sp.symbols('x lambda')
y = sp.Function('y')(x)

ode = sp.Eq(y.diff(x, x) - lam * y, 0) #Ecuacion diferencial

sol = sp.dsolve(ode, y) #Solucion

C1, C2 = sp.symbols('C1 C2') #Constantes
y_general = sol.rhs.subs({sol.rhs.args[0].args[0]: C1, sol.rhs.args[1].args[0]: C2})  #Valores de fronteras

eq1 = y_general.subs(x, 0) #X(0)=0
C1_val = sp.solve(eq1, C1)[0]

y_general = y_general.subs(C1, C1_val)  #Nueva solucion despues de X(0)=0

eq2 = y_general.subs(x, 1) #X(1)=0
lambda_vals = sp.solve(eq2, lam)  # Valores de lambda

lambda_n = [(n * sp.pi) ** 2 for n in range(1, 5)]  #Extraer valores de lambda permitidos (nπ)^2. Primeros 4 valores

y_vals = np.linspace(0, 1, 100) # Graficar las primeras soluciones propias
plt.figure(figsize=(8, 5))

for n, lam_n in enumerate(lambda_n, 1):
    y_n = sp.sin(n * sp.pi * x)  # Función propia
    y_n_func = sp.lambdify(x, y_n, 'numpy')  # Convertir a función NumPy
    plt.plot(y_vals, y_n_func(y_vals), label=f"n={n}")

plt.title("Primeras soluciones propias de X'' - λX = 0")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.grid()
plt.show()

```

$X^{''}-\lambda X=0 \qquad X'(0)=X'(1)=0$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, lam = sp.symbols('x lambda')
y = sp.Function('y')(x)

ode = sp.Eq(y.diff(x, x) - lam * y, 0) #Ecuacion diferencial

sol = sp.dsolve(ode, y) #Solucion

C1, C2 = sp.symbols('C1 C2') #Constantes
y_general = sol.rhs.subs({sol.rhs.args[0].args[0]: C1, sol.rhs.args[1].args[0]: C2})  #Valores de fronteras

y_derivative = sp.diff(y_general, x) # Calcular la derivada X'(x)

eq1 = y_derivative.subs(x, 0) #X'(0)=0
C2_val = sp.solve(eq1, C2)[0]

y_general = y_general.subs(C2, C2_val) # Nueva solución después de aplicar X'(0) = 0

eq2 = y_derivative.subs(x, 1) #X'(1)=0
lambda_vals = sp.solve(eq2, lam)  # Valores de lambda

lambda_n = [(n * sp.pi) ** 2 for n in range(1, 5)]  #Extraer valores de lambda permitidos (nπ)^2. Primeros 4 valores

y_vals = np.linspace(0, 1, 100) # Graficar las primeras soluciones propias
plt.figure(figsize=(8, 5))

for n, lam_n in enumerate(lambda_n, 1):
    y_n = sp.cos(n * sp.pi * x / 2)  # Función propia
    y_n_func = sp.lambdify(x, y_n, 'numpy')  # Convertir a función NumPy
    plt.plot(y_vals, y_n_func(y_vals), label=f"n={n}")

plt.title("Primeras soluciones propias de X'' - λX = 0 con X'(0) = X'(1) = 0")
plt.xlabel("x")
plt.ylabel("X(x)")
plt.legend()
plt.grid()
plt.show()
```
