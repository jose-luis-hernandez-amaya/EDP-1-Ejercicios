### Curvas parametrizadas

Se importa para graficar curvas paramétricas



```python
from sympy import plot_parametric
```

Define una variable simbólica

```python
s=symbols('s')
```

Grafica la curva paramétrica $x=cos(s),y=sen(s)$, que representa un círculo unitario

```python
plot_parametric((cos(s)),sin(s))
```

Se grafican 2 funciones parametricas, un ciruclo unitario y una parabola, de [-4,4] en el eje X

```python
plot_parametric((cos(s),sin(s)),(s,s**2),(s,-4,4))
```

### Graficas 3D

Se importa para hacer funciones en <font color='red'>3 dimensiones</font>

```python
from sympy.plotting import plot3d
```

Se definen 2 variables simbolicas <font color='red'>x</font> y <font color='red'>y</font>

```python
x,y=symbols('x y')
```

Grafica la funcion $f(x,y)= x*y$ con <font color='red'>plot3d</font>

```python
plot3d(x*y)
```

Grafica la misma funcion $f(x,y)=x*y$, pero desde [-5,5] en el eje X y Y

```python
plot3d(x*y, (x,-5,5),(y,-5,5))
```

Grafica 2 funciones $f(x,y)=x*y$ y $f(x,y)=-x*y$, desde [-5,5] en el eje X y Y

```python
plot3d(x*y, -x*y, (x,-5,5),(y,-5,5))
```

Se grafican 2 funciones, $f(x,y)=x^2+y^2$ desde [-5,5] en el eje X y Y, y $f(x,y)=x*y$ desde [-3,3] en el eje X y Y

```python
plot3d((x**2 + y**2, (x,-5,5),(y,-5,5)),(x*y,(x,-3,3),(y,-3,3)))
```

Se usa para graficar curvas paramétricas en 3D solamente la pura <font color='red'>linea</font>

```python
from sympy.plotting import plot3d_parametric_line
```

Utilizo la misma variable simbolica que se habia creado antes y se grafica un tipo de resorte con funciones $sen(s)$ y $cos(s)$

```python
plot3d_parametric_line(cos(s),sin(s),s)
```

Se grafican 2 funciones parametricas, una con $x=cos(s),y=sen(s),z=s$ con rango de [-5,5] en s, y la funcion $x=sen(s),y=s^2,z=s$ con rango de [-5,5] en s

```python
plot3d_parametric_line((cos(s),sin(s),s,(s,-5,5)),(sin(s),s**2,s,(s,-5,5)))
```

Se usa para graficar curvas paramétricas en 3D "completas"

```python
from sympy.plotting import plot3d_parametric_surface
```

Se crean 2 variables simbolicas <font color='red'>u</font> y <font color='red'>v</font>

```python
u, v =symbols('u v')
```

Se grafica la funcion parametrica $x=cos(u+v), y=sen(u-v),z=u-v$

```python
plot3d_parametric_surface(cos(u+v),sin(u-v),u-v)
```

Se grafica la funcion parametrica $x=cos(v)*cos(u), y=cos(v)*sen(u),z=sen(v)$

```python
plot3d_parametric_surface(cos(v)*cos(u),cos(v)*sin(u),sin(v))
```

Funciones implícitas

 Sirve para graficar ecuaciones e inecuaciones implícitas en 2D

```python
from sympy import plot_implicit, symbols, Eq, And
```

Se crean 2 variables simbolicas <font color='red'>x</font> y <font color='red'>y</font>

```python
x, y = symbols(' x y')
```

Es una ecuacion de la circunferencia $x^2+y^2=5^2$, utilizando <font color='red'>plot_implicit</font> y <font color='red'>Eq</font> de Ecuaciones

```python
p1 = plot_implicit(Eq(x**2 + y**2, 5))
```

Grafica una region de la funcion $f> x^2$, con <font color='red'>plot_implicit</font>

```python
p6 = plot_implicit(y > x**2)
```

Genera una funcion $y(x)=|x|$, solo que esta descrita con el $AND$ que une las 2 condiciones $y>x$ y $y>-x$

```python
p6 = plot_implicit(And(y > x, y > -x))
```
