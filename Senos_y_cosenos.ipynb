```python
from sympy import fourier_series, cos, pi, exp, plot
from sympy.abc import x

# Definir la función f(x) = e^(2x)
f = exp(2*x)

# Calcular la serie de Fourier en cosenos de f(x) en el intervalo [0, 1]
s = fourier_series(f, (x, -1, 1))

# Truncar la serie a diferentes números de términos para la visualización
s1 = s.truncate(n=3)  # Serie truncada con 3 términos
s2 = s.truncate(n=5)  # Serie truncada con 5 términos
s3 = s.truncate(n=7)  # Serie truncada con 7 términos

# Graficar la función original y las aproximaciones de la serie de Fourier
p = plot(f, s1, s2, s3, (x, -1, 1), show=False, legend=True)
p[0].label = 'f(x) = exp(2x)'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
p.show()

# Imprimir la serie de Fourier truncada
print("Serie de Fourier truncada (n=7):")
s3
```

```python

from sympy import fourier_series, sin, pi, exp, plot
from sympy.abc import x

# Definir la función f(x) = e^(2x)
f = exp(2*x)

# Calcular la serie de Fourier en senos de f(x) en el intervalo [0, 1]
s = fourier_series(f, (x, -1, 1), (sin,)) # Se especifica (sin,) para la serie de Fourier en senos

# Truncar la serie a diferentes números de términos para la visualización
s1 = s.truncate(n=3)  # Serie truncada con 3 términos
s2 = s.truncate(n=5)  # Serie truncada con 5 términos
s3 = s.truncate(n=7)  # Serie truncada con 7 términos

# Graficar la función original y las aproximaciones de la serie de Fourier
p = plot(f, s1, s2, s3, (x, -1, 1), show=False, legend=True)
p[0].label = 'f(x) = exp(2x)'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
p.show()

# Imprimir la serie de Fourier truncada (opcional)
print("Serie de Fourier en senos truncada (n=7):")
s3
```

```python

from sympy import fourier_series, cos, sin, pi, Piecewise, plot
from sympy.abc import x

# Definir la función a trozos
f = Piecewise((1, (x >= 0) & (x <= pi/2)), (2, (x > pi/2) & (x <= pi)))

# Calcular la serie de Fourier
s = fourier_series(f, (x, 0, pi))

# Truncar la serie para diferentes valores de n
s1 = s.truncate(n=3)
s2 = s.truncate(n=5)
s3 = s.truncate(n=7)

# Graficar la función original y las aproximaciones
p = plot(f, s1, s2, s3, (x, 0, pi), show=False, legend=True)

p[0].label = 'f(x)'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'

p.show()

# Imprimir la serie de Fourier truncada
print("Serie de Fourier (n=3):")
print(s1)
print("\nSerie de Fourier (n=5):")
print(s2)
print("\nSerie de Fourier (n=7):")
s3
```
