```python
# prompt: Programa un codigo que de una grafica de y=x^2

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)  # Crea 100 puntos entre -5 y 5
y = x**2

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.title("Gr√°fica de y = x^2")
plt.grid(True)
plt.show()
```
