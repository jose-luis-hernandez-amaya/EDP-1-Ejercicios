Grafique la funcion $$f(x')=exp(-\frac{(x'-0.5)^2}{0.01})$$
Y suponga que esta representa una pertubacion en una cuerda

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-((x - 0.5)**2) / 0.01)

x = np.linspace(0, 1, 400)
y = f(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label=r'$f(x^\prime) = \exp\left(-\frac{(x^\prime - 0.5)^2}{0.01}\right)$')
plt.title('Perturbaci√≥n en una cuerda')
plt.xlabel(r'$x^\prime$')
plt.ylabel(r'$f(x^\prime)$')
plt.grid(True)
plt.legend()
plt.show()
```

a) Mostrar que al hacer el cambio de variable $$x'=x-ct$$
la pertubacion se desplaza a la derecha

```python
import numpy as np
import matplotlib.pyplot as plt

c = 1
t_vals = [0, 0.1, 0.2, 0.3]

def f_right(x, t, c):
    x_prime = x - c * t
    return np.exp(-((x_prime - 0.5)**2) / 0.01)

x = np.linspace(0, 1.5, 500)

plt.figure(figsize=(10, 5))
for t in t_vals:
    y = f_right(x, t, c)
    plt.plot(x, y, label=f't = {t}')

plt.title('Onda viajera a la derecha: $f(x - ct)$')
plt.xlabel('x')
plt.ylabel('f(x - ct)')
plt.grid(True)
plt.legend()
plt.show()
```

b) Mostrar que al hacer el cambio de variable $$x'=x+ct$$
la pertubacion se desplaza a la izquierda

```python
import numpy as np
import matplotlib.pyplot as plt

c = 1
t_vals = [0, 0.1, 0.2, 0.3]

def f_left(x, t, c):
    x_prime = x + c * t
    return np.exp(-((x_prime - 0.5)**2) / 0.01)

x = np.linspace(0, 1.5, 500)

plt.figure(figsize=(10, 5))
for t in t_vals:
    y = f_left(x, t, c)
    plt.plot(x, y, label=f't = {t}')

plt.title('Onda viajera a la izquierda: $f(x + ct)$')
plt.xlabel('x')
plt.ylabel('f(x + ct)')
plt.grid(True)
plt.legend()
plt.show()
```

c) Graficar la superposici√≥n de las ondas viajeras de los incisos a) y b)

```python
plt.figure(figsize=(10, 5))
for t in t_vals:
    y = f_left(x, t, c)
    plt.plot(x, y, label=f't = {t}')
    y = f_right(x, t, c)
    plt.plot(x, y, label=f't = {t}')

plt.title("Onda viajera con $x'$ =x-ct y $x'$ =x+ct" )
plt.xlabel('x')
plt.ylabel('f(x +- ct)')
plt.grid(True)
plt.show()
```
