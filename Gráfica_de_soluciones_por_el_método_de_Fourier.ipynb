```python
from pylab import *

def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = x[i]
    return y

x = linspace(0,1,1000)
f = recta(x)

def u_0(x,N=20): # N es el nmero de armnicos
    f = zeros(len(x))
    for n in range(1,N):
        f += 2*(((-1)**n - 1)/(n**2 * pi**2))*(e**(-n**2*pi**2*0))*cos(n*pi*x)  #Aqui t vale cero
    return 1/2 + f

f_n = u_0(x)
plot(x,f_n,label='$f_N(x)$ con $N=20$')
plot(x,f,label='$f(x)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximacin de la condicin inicial")
xlim([-0.01,1.1]),ylim([-0.01,1.2]),grid(),legend()
show()

#Ahora con t
def u_x_t(x,t,N=50):
    f = zeros( (len(x),len(t)) )
    for i in range(len(t)):
        for n in range(1,N):
            f[:,i] += 2*(((-1)**n - 1)/(n**2 * pi**2))*(e**(-n**2 * pi**2 *t[i]))*cos(n*pi*x)
    return 1/2+f

x = linspace(0,1,1000)
t = linspace(0,10,1000)
U = u_x_t(x,t)

plot(x,U[:,0],label='$u(x,0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximacin de la condicin inicial")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()

plot(x,U[:,0],label='$u(x,t=0)$') # Mejorar
plot(x,U[:,1],label='$u(x,t=1)$')
plot(x,U[:,2],label='$u(x,t=2)$')
plot(x,U[:,3],label='$u(x,t=3)$')
plot(x,U[:,10],label='$u(x,t=10)$')
plot(x,U[:,20],label='$u(x,t=20)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,t)$');
plt.title("Evolucin de la solucin")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,100)
t = linspace(0,1.0,100)

X,T = np.meshgrid(x,t)

U = u_x_t(x,t)

plt.xlabel('$t$'); plt.ylabel('$x$');
plt.title("u(x,t)")
ax.plot_surface(X, T, U, cmap="inferno"); #Otros colores:"viridis", "inferno", "plasma"
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funcin u(x,t)
def u_x_t(x, t):
    return np.sin(np.pi * x) * np.cos(np.pi * t)

# Crear el grfico
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection="3d")

# Crear los datos para x y t
x = np.linspace(0, 1.0, 100)
t = np.linspace(0, 1.0, 100)

X, T = np.meshgrid(x, t)

# Evaluar la funcin u(x,t)
U = u_x_t(X, T)

# Etiquetas y ttulo
plt.xlabel('$t$')
plt.ylabel('$x$')
plt.title("u(x,t)")

# Graficar la superficie
ax.plot_surface(X, T, U, cmap="inferno")  # Otros colores: "viridis", "inferno", "plasma"

# Mostrar el grfico
plt.show()
```
